```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("employee_db.db")
cursor = conn.cursor()

print("Database Connected Successfully")
```

```python
cursor.execute("""
CREATE TABLE Employees (
    EmpID INTEGER PRIMARY KEY,
    EmpName TEXT,
    Department TEXT,
    City TEXT,
    Salary INTEGER,
    HireDate DATE
)
""")

print("Employees Table Created Successfully")
```

#Question 1
Show employees working in either the IT or HR departments.

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE Department IN ('IT','HR')
""", conn)

df
```

#Question 2
Retrieve employees whose department is Sales, IT or Finance.

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE Department IN ('Sales','IT','Finance')
""", conn)

df
```

#Question 3
Display employees whose salary is between ₹50,000 and ₹70,000.

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE Salary BETWEEN 50000 AND 70000
""", conn)

df
```

#Question 4
List employees whose names start with the letter A.

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE EmpName LIKE 'A%'
""", conn)

df
```

#Question 5
Find employees whose names contain the substring "an".

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE EmpName LIKE '%an%'
""", conn)

df
```

#Question 6
Show employees who are from Delhi or Mumbai and earn more than ₹55,000.

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE City IN ('Delhi','Mumbai')
AND Salary > 55000
""", conn)

df
```

#Question 7
Display all employees except those from the HR department.

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE Department <> 'HR'
""", conn)

df
```

#Question 8
Get all employees hired between 2019 and 2022 ordered by HireDate (oldest first).

```python
df = pd.read_sql_query("""
SELECT *
FROM Employees
WHERE HireDate BETWEEN '2019-01-01' AND '2022-12-31'
ORDER BY HireDate ASC
""", conn)

df
```
